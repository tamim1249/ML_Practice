```python
Vocabulary_initial=['A','B','C','D']
word_2024=[500,800,780,900]
word_2025=[820,999,236,983]
x=np.arange(len(Vocabulary_initial))
plt.bar(x-0.2,word_2024,color='purple',label='2024',width=0.5,edgecolor='black')
plt.bar(x+0.2,word_2025,color='green',width=0.5,edgecolor='red',label='2025')
plt.xlabel('Vocabulary_initial',color='green')
plt.ylabel('word count year',color='orange')
plt.title("Memorize vocabulary per year",color='red')
plt.legend(loc='best')
plt.xticks(x,Vocabulary_initial)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

student = ['Tamim', 'Rabbi', 'Ridoy', 'Nadim', 'Meer']
cgpa = [2.90, 3.77, 3.00, 3.80, 2.18]
skill = [3.90, 3.50, 3.40, 3.35, 3.80]

x = np.arange(len(student))

plt.bar(x - 0.2, cgpa, color='mediumseagreen', width=0.4, label='CGPA',edgecolor='black')
plt.bar(x + 0.2, skill, color='coral', width=0.4, label='Skill',edgecolor='red')

plt.xticks(x, student)  # x-axis labels
plt.title("CGPA vs Skill Comparison")
plt.xlabel("Students", color='blue')
plt.ylabel("Score")
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
